```python
import numpy as np
import matplotlib.pyplot as plt
from spiking_radio_reservoir import setup_hennequin_connectivity, setup_input_layer, setup_reservoir_layer
from utils.reservoir import getTauCurrent, getTau
from utils.channel import send
from utils.modulator import AsynchronousDeltaModulator, modulate
from utils.plotting import plot_sample_with_labels
from brian2 import us, ms, second, amp, defaultclock
from teili import TeiliNetwork
```

```python
params = {
    'wGen': 8000,
    'wInp': 5000,
    'pIR': 0.1,
    'pE_local': 0.5,
    'pI_local': 1.0,
    'k': 3,
    'DoC': 0.2,
    'loc_wResE': 8000,
    'scale_wResE': 100,
    'loc_wResI': -8000,
    'scale_wResI': 100,
    'N': 200,
    'tau': 20,
    'Ngx': 10,
    'Ngy': 10,
}
```

```python
connectivity = setup_hennequin_connectivity(params['N'], params['pIR'], params['Ngx'], params['Ngy'], \
    params['pE_local'], params['pI_local'], params['k'], params['DoC'], \
    params['loc_wResE'], params['scale_wResE'], params['loc_wResI'], params['scale_wResI'])
```

```python
input_neurons_I = connectivity['inp_res']['j'][connectivity['inp_res']['i']==0]
input_neurons_Q = connectivity['inp_res']['j'][connectivity['inp_res']['i']==1]
M = params['Ngx']*params['Ngy']
colors = []
for n in range(params['N']):
    if n in input_neurons_I:
        colors.append('#2ecc71')
    elif n in input_neurons_Q:
        colors.append('#e67e22')
    elif n < M:
        colors.append('#e74c3c')
    else:
        colors.append('#3498db')
```

```python
input_string = "neuro"
time_sample, sample, time_labels, labels, bins = send(input_string, m=8)
settings = {
    'thrup': 0.1,
    'thrdn': 0.1,
    'resampling_factor': 10
}
modulators = [
    AsynchronousDeltaModulator(settings['thrup'], settings['thrdn'], settings['resampling_factor']),
    AsynchronousDeltaModulator(settings['thrup'], settings['thrdn'], settings['resampling_factor'])
]
indices, times, time_stim, signal = modulate(modulators[0], modulators[1], time_sample, sample, resampling_factor=settings['resampling_factor'])
plot_sample_with_labels(signal, time_stim, indices, times, labels, time_labels, bins, figsize=(12, 9))
```

```python
components = {'generator': None, 'layers': None, 'synapsis': None, 'monitors': None}
components = setup_input_layer(components, params['wGen'])
Itau = getTauCurrent(2*ms)
components = setup_reservoir_layer(components, connectivity, params['N'], Itau, params['wInp'])
components['generator'].set_spikes(indices, times*second)
```

```python
defaultclock.dt = 5*us
network = TeiliNetwork()
network.add(components['generator'])
network.add(list(components['layers'].values()))
network.add(list(components['synapsis'].values()))
network.add(list(components['monitors'].values()))
```

```python
network.run(40*ms)
```

```python
positions = []
for n in range(params['N']):
    idx = np.where(network['mRes'].i==n)[0]
    positions.append(network['mRes'].t[idx]/ms)
```

```python
plt.figure(figsize=(12, 9))
plt.eventplot(positions, colors=colors)
plt.ylabel("neuron index")
plt.xlabel("time [ms]")
```
