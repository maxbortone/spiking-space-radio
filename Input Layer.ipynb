```python
import joblib
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("./connectivity-hennequin.joblib", "rb") as fo:
    connectivity = joblib.load(fo)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 9))
ax1.scatter(connectivity['gen_inp']['i'], connectivity['gen_inp']['j'], c=connectivity['gen_inp']['w'], marker='.')
ax1.set_xlabel('spike generator')
ax1.set_ylabel('target neuron')
ax1.tick_params(direction='in')
ax1.set_xticklabels(['I.up', 'I.dn', 'Q.up', 'Q.dn'])
ax2.scatter(connectivity['inp_res']['i'], connectivity['inp_res']['j'], c='k', marker='.')
ax2.set_xlabel('input neuron')
ax2.tick_params(direction='in')
ax2.yaxis.tick_right()
ax2.yaxis.set_label_position("right")
ax2.yaxis.set_major_locator(plt.MultipleLocator(25))
ax2.set_yticklabels([])
```

```python
generator_neurons = [(0, i) for i in range(8)]
input_neurons = [(1, i) for i in range(4)]
colors = ["#3498db", "#e74c3c"]
synapses = []
for i in range(4):
    for j in range(2):
        pre = generator_neurons[i*2+j]
        post = input_neurons[i]
        synapses.append([pre, post, colors[j]])
```

```python
fig, ax = plt.subplots(figsize=(9, 6))
ax.scatter(np.zeros(8), np.arange(8), color="k")
ax.scatter(np.ones(4), np.arange(4)+2, color="k")
for s in synapses:
    x = [s[0][0], s[1][0]]
    y = [s[0][1], s[1][1]+2]
    ax.plot(x, y, c=s[2], linewidth=1, zorder=0)
ax.set_xticks([0, 1])
ax.set_xticklabels(["generator", "input"], fontsize=18)
ax.set_yticks(np.arange(8))
ax.set_yticklabels(["Q.up", "Q.dn", "Q.dn", "Q.up", "I.up", "I.dn", "I.dn", "I.up"], fontsize=18)
plt.savefig("images/input-layer.pdf", bbox_inches="tight")
```
